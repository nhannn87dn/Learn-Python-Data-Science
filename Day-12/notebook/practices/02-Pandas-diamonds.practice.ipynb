**Giải thích trường dữ liệu**

| Cột (field) | Ý nghĩa                                                                                   | Đơn vị / Giá trị điển hình                                      | Ghi chú quan trọng                                                                 |
|-------------|-------------------------------------------------------------------------------------------|-----------------------------------------------------------------|-----------------------------------------------------------------------------------|
| **carat**   | Trọng lượng kim cương                                                                     | 0.2 → 5.01 (thường tập trung 0.3–1.5)                           | 1 carat = 0.2 gram. Carat là yếu tố ảnh hưởng lớn nhất đến giá (giá tăng gần theo hàm mũ với carat). |
| **cut**     | Chất lượng cắt (tỷ lệ, độ đối xứng, độ đánh bóng)                                         | Ideal > Premium > Very Good > Good > Fair                       | Ideal là đẹp nhất (ánh sáng phản xạ tốt nhất) → giá cao hơn cùng các yếu tố khác. |
| **color**   | Màu sắc kim cương (độ không màu)                                                          | D (tốt nhất, hoàn toàn không màu) → J (tệ nhất, có sắc vàng nhẹ) | Thang màu chuẩn GIA: D-E-F (không màu), G-H-I-J (gần không màu), K trở đi có màu rõ. D là hiếm và đắt nhất. |
| **clarity** | Độ trong suốt (mức độ không có vết/vết nhỏ bên trong)                                     | IF > VVS1 > VVS2 > VS1 > VS2 > SI1 > SI2 > I1 (xấu nhất)          | IF = Flawless (không tỳ vết hoàn hảo), VVS = rất rất nhỏ, VS = rất nhỏ, SI = nhỏ, I = có vết nhìn thấy được. |
| **depth**   | Tỷ lệ độ sâu tổng thể của viên kim cương                                                  | % (thường 58–66%)                                               | Công thức: depth % = (z / trung bình(x,y)) × 100. Khoảng 60–62.5% thường được đánh giá là lý tưởng. |
| **table**   | Tỷ lệ chiều rộng mặt bàn (mặt trên phẳng) so với đường kính rộng nhất của viên kim cương  | % (thường 52–66%)                                               | Khoảng 54–60% thường cho ánh sáng tốt nhất. Quá rộng hoặc quá hẹp đều làm giảm độ sáng. |
| **price**   | Giá bán của viên kim cương                                                                | USD (326 → 18.823 USD trong dataset này)                        | Biến mục tiêu khi dự đoán giá. Giá tăng rất mạnh theo carat và chất lượng.        |
| **x**       | Chiều dài viên kim cương                                                                  | mm (0 → 10.74 mm)                                               |                                                                                   |
| **y**       | Chiều rộng viên kim cương                                                                 | mm (0 → 58.9 mm – có vài giá trị ngoại lai rất lớn)              |                                                                                   |
| **z**       | Chiều sâu (độ cao) viên kim cương                                                         | mm (0 → 31.8 mm)                                                | z = 0 là lỗi dữ liệu (có vài hàng như vậy trong dataset gốc).                     |
