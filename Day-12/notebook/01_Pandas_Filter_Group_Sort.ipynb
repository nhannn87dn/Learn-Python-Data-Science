## üîç II. L·ªçc d·ªØ li·ªáu n√¢ng cao
---

### L·ªçc b·∫±ng `boolean mask`

Boolean mask l√† c√°ch t·∫°o **bi·ªÉu th·ª©c ƒëi·ªÅu ki·ªán tr·∫£ v·ªÅ True/False** ƒë·ªÉ l·ªçc d√≤ng d·ªØ li·ªáu.

```python
# Doanh thu l·ªõn h∆°n 5000$ ?
df[df['Sales'] > 5000]
```

```python
# Giao d·ªãch c·ªßa Vietnam qua k√™nh Online
df[(df['Country'] == 'Vietnam') & (df['SalesChannel'] == 'Online')]
```

```python
# T√¨m l·ª£i nhu·∫≠n c·ªßa nh√≥m h√†ng Electronics l·ªõn 800$
df[(df['Category'] == 'Electronics') & (df['Profit'] > 800)]
```

### L·ªçc b·∫±ng `query()`

```python
# Doanh thu l·ªõn h∆°n 5000$
df.query('Sales > 5000')
```

```python
# Giao d·ªãch c·ªßa Vietnam qua k√™nh Online
df.query('Country == "Vietnam" and SalesChannel == "Online"')
```

```python
# T√¨m l·ª£i nhu·∫≠n c·ªßa nh√≥m h√†ng Electronics l·ªõn 800$
threshold = 800
df.query('Profit > @threshold and Category == "Electronics"')
```

```python
# M·ªôt s·ªë v√≠ d·ª• n√¢ng cao
df.query('Country in ["Vietnam", "Thailand"]')
```

```python
# T√™n s·∫£n ph·∫©m c√≥ ch·ª©a k√≠ t·ª± 'a'
df[df['Product'].str.contains('a', case=False)]
```

```python
#T√¨m theo kho·∫£ng ng√†y
df['Date'] = pd.to_datetime(df['Date'])
df.query('"2024-01-05" <= Date <= "2024-01-10"')
```

## üìä III. Nh√≥m d·ªØ li·ªáu v√† t√≠nh to√°n t·ªïng h·ª£p
---

### Nh√≥m d·ªØ li·ªáu `groupby`

```python
# V√≠ d·ª• 1: T√≠nh t·ªïng doanh thu theo qu·ªëc gia
df.groupby('Country')['Sales'] # nh√≥m
df.groupby('Country')['Sales'].sum()
```

```python
# V√≠ d·ª• 2: T√≠nh t·ªïng l·ª£i nh·∫≠n theo qu·ªëc gia
df.groupby('Country')['Profit']
df.groupby('Country')['Profit'].sum()
```

### T√≠nh to√°n t√™n nh√≥m `aggregate`

```python
# Nhi·ªÅu ph√©p t√≠nh c√πng l√∫c
df.groupby('Country').agg({
    'Sales': ['sum', 'mean'],
    'Profit': ['sum', 'mean', 'max']
})
```

```python
# Nh√≥m theo nhi·ªÅu c·ªôt
df.groupby(['Country', 'Category']).agg({
    'Sales': 'sum',
    'Profit': 'mean'
})
```

```python
# ƒê·∫øm s·ªë ƒë∆°n h√†ng v√† t·ªïng s·ªë l∆∞·ª£ng b√°n
df.groupby('Category').agg(
    Orders=('OrderID', 'count'),
    TotalQuantity=('Quantity', 'sum')
)
```

```python
# L·ªçc nh√≥m sau khi group
df.groupby('Country').filter(lambda x: x['Sales'].sum() > 5000)
```

```python
# S·∫Øp x·∫øp theo gi√° tr·ªã doanh thu tƒÉng d·∫ßn
df.sort_values(by='Sales', ascending=True)
```

```python
# S·∫Øp x·∫øp theo gi√° tr·ªã l·ª£i nh·∫≠n gi·∫£m d·∫ßn
df.sort_values(by='Profit', ascending=False)
```

```python
# S·∫Øp x·∫øp nhi·ªÅu tr∆∞·ªùng
df.sort_values(by=['Country', 'Sales'], ascending=[True, False])
```

```python
# S·∫Øp x·∫øp theo ch·ªâ m·ª•c
df = df.set_index('Date')
df.sort_index(ascending=True)
```
