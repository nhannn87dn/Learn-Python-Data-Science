## üß© 1. Iterating Arrays (Duy·ªát ph·∫ßn t·ª≠ trong m·∫£ng)

### üß† ƒê·ªãnh nghƒ©a:
Duy·ªát m·∫£ng (iteration) nghƒ©a l√† **truy c·∫≠p t·ª´ng ph·∫ßn t·ª≠ trong m·∫£ng NumPy**.

C√≥ th·ªÉ duy·ªát m·∫£ng b·∫±ng **v√≤ng l·∫∑p `for`**, ho·∫∑c **d√πng `np.nditer()`** khi l√†m vi·ªác v·ªõi m·∫£ng nhi·ªÅu chi·ªÅu.

---
### üìò V√≠ d·ª• 1: Duy·ªát m·∫£ng 1 chi·ªÅu

```python
import numpy as np

arr = np.array([10, 20, 30])
for x in arr:
    print(x)
```

### üìò V√≠ d·ª• 2: Duy·ªát m·∫£ng 2 chi·ªÅu th√¥ng th∆∞·ªùng

```python
arr2d = np.array([[1, 2, 3], [4, 5, 6]])

for row in arr2d:
    for val in row:
        print(val, end=" ")
```

### üìò V√≠ d·ª• 3: Duy·ªát m·∫£ng nhi·ªÅu chi·ªÅu b·∫±ng `np.nditer()`

```python
arr3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

for x in np.nditer(arr3d):
    print(x)
```

## üîó 2. Joining Arrays (Gh√©p m·∫£ng)

### üß† ƒê·ªãnh nghƒ©a:
Gh√©p m·∫£ng l√† **k·∫øt h·ª£p nhi·ªÅu m·∫£ng l·∫°i th√†nh m·ªôt m·∫£ng l·ªõn h∆°n**.

### ‚öôÔ∏è M·ªôt s·ªë h√†m ph·ªï bi·∫øn:
- `np.concatenate((a1, a2), axis=0)` ‚Äî gh√©p theo tr·ª•c (h√†ng ho·∫∑c c·ªôt)
- `np.stack((a1, a2), axis=0)` ‚Äî th√™m tr·ª•c m·ªõi khi gh√©p
- `np.hstack((a1, a2))` ‚Äî gh√©p theo chi·ªÅu ngang
- `np.vstack((a1, a2))` ‚Äî gh√©p theo chi·ªÅu d·ªçc

---
### üìò V√≠ d·ª• 1: D√πng `np.concatenate()` v·ªõi m·∫£ng 1D

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

result = np.concatenate((arr1, arr2))
print(result)
```

### üìò V√≠ d·ª• 2: D√πng `np.concatenate()` v·ªõi m·∫£ng 2D

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

print("Gh√©p theo h√†ng (axis=0):\n", np.concatenate((a, b), axis=0))
print("\nGh√©p theo c·ªôt (axis=1):\n", np.concatenate((a, b), axis=1))
```

### üìò V√≠ d·ª• 3: D√πng `np.stack()` ƒë·ªÉ th√™m tr·ª•c m·ªõi

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

stacked = np.stack((arr1, arr2), axis=1)
print(stacked)
```

### üìò V√≠ d·ª• 4: Gh√©p ngang v√† d·ªçc b·∫±ng `hstack()` v√† `vstack()`

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Gh√©p ngang:", np.hstack((a, b)))
print("Gh√©p d·ªçc:\n", np.vstack((a, b)))
```

## ‚úÇÔ∏è 3. Splitting Arrays (T√°ch m·∫£ng)

### üß† ƒê·ªãnh nghƒ©a:
T√°ch m·∫£ng l√† **chia m·ªôt m·∫£ng l·ªõn th√†nh nhi·ªÅu m·∫£ng nh·ªè**.

### ‚öôÔ∏è C√°c h√†m ph·ªï bi·∫øn:
- `np.split(array, n)` ‚Äî t√°ch ƒë·ªÅu m·∫£ng th√†nh *n ph·∫ßn*.
- `np.array_split(array, n)` ‚Äî t√°ch kh√¥ng ƒë·ªÅu n·∫øu kh√¥ng chia h·∫øt.
- `np.hsplit()` ‚Äî t√°ch theo chi·ªÅu ngang (c·ªôt).
- `np.vsplit()` ‚Äî t√°ch theo chi·ªÅu d·ªçc (h√†ng).

---
### üìò V√≠ d·ª• 1: D√πng `np.split()` ƒë·ªÉ chia ƒë·ªÅu m·∫£ng 1D

```python
arr = np.array([1, 2, 3, 4, 5, 6])

x = np.split(arr, 3)
print(x)
```

### üìò V√≠ d·ª• 2: D√πng `np.array_split()` ƒë·ªÉ t√°ch kh√¥ng ƒë·ªÅu

```python
arr = np.array([1, 2, 3, 4, 5, 6, 7])
x = np.array_split(arr, 3)
print(x)
```

### üìò V√≠ d·ª• 3: D√πng `np.hsplit()` v√† `np.vsplit()` v·ªõi m·∫£ng 2D

```python
arr2d = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])

print("T√°ch theo c·ªôt (hsplit):", np.hsplit(arr2d, 2))
print("\nT√°ch theo h√†ng (vsplit):", np.vsplit(arr2d, 2))
```

- **Joining** gi√∫p g·ªôp m·∫£ng d·ªÖ d√†ng b·∫±ng `concatenate`, `stack`, `hstack`, `vstack`.
- **Splitting** h·ªó tr·ª£ chia nh·ªè m·∫£ng b·∫±ng `split`, `array_split`, `hsplit`, `vsplit`.
