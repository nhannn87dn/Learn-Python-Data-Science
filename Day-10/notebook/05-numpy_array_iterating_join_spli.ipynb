## ğŸ§© 1. Iterating Arrays (Duyá»‡t pháº§n tá»­ trong máº£ng)

### ğŸ§  Äá»‹nh nghÄ©a:
Duyá»‡t máº£ng (iteration) nghÄ©a lÃ  **truy cáº­p tá»«ng pháº§n tá»­ trong máº£ng NumPy**.

CÃ³ thá»ƒ duyá»‡t máº£ng báº±ng **vÃ²ng láº·p `for`**, hoáº·c **dÃ¹ng `np.nditer()`** khi lÃ m viá»‡c vá»›i máº£ng nhiá»u chiá»u.

### ğŸ“˜ VÃ­ dá»¥ 1: Duyá»‡t máº£ng 1 chiá»u

```python
import numpy as np

arr = np.array([10, 20, 30])
for x in arr:
    print(x)
```

### ğŸ“˜ VÃ­ dá»¥ 2: Duyá»‡t máº£ng 2 chiá»u thÃ´ng thÆ°á»ng

```python
arr2d = np.array([[1, 2, 3], [4, 5, 6]])

for row in arr2d:
    for val in row:
        print(val, end=" ")
```

### ğŸ“˜ VÃ­ dá»¥ 3: Duyá»‡t máº£ng nhiá»u chiá»u báº±ng `np.nditer()`

```python
arr3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

for x in np.nditer(arr3d):
    print(x)
```

## ğŸ”— 2. Joining Arrays (GhÃ©p máº£ng)

### ğŸ§  Äá»‹nh nghÄ©a:
GhÃ©p máº£ng lÃ  **káº¿t há»£p nhiá»u máº£ng láº¡i thÃ nh má»™t máº£ng lá»›n hÆ¡n**.

### âš™ï¸ Má»™t sá»‘ hÃ m phá»• biáº¿n:
- `np.concatenate((a1, a2), axis=0)` â€” ghÃ©p theo trá»¥c (hÃ ng hoáº·c cá»™t)
- `np.stack((a1, a2), axis=0)` â€” thÃªm trá»¥c má»›i khi ghÃ©p
- `np.hstack((a1, a2))` â€” ghÃ©p theo chiá»u ngang
- `np.vstack((a1, a2))` â€” ghÃ©p theo chiá»u dá»c

Trong NumPy, **trá»¥c (axis)** Ä‘Æ¡n giáº£n lÃ  **chiá»u** hoáº·c **hÆ°á»›ng** cá»§a máº£ng (array). Má»—i máº£ng cÃ³ thá»ƒ cÃ³ nhiá»u chiá»u, vÃ  má»—i chiá»u Ä‘Æ°á»£c Ä‘Ã¡nh sá»‘ bá»Ÿi má»™t axis.

ğŸ‘‰ Hiá»ƒu Ä‘Æ¡n giáº£n:

- axis = 0: chiá»u dá»c (xuá»‘ng dÆ°á»›i)
- axis = 1: chiá»u ngang (sang pháº£i)
- axis = 2 trá»Ÿ lÃªn: chiá»u sÃ¢u, chiá»u thá»© 3, 4,...

### ğŸ“˜ VÃ­ dá»¥ 1: DÃ¹ng `np.concatenate()` vá»›i máº£ng 1D

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

result = np.concatenate((arr1, arr2))
print(result)
```

### ğŸ“˜ VÃ­ dá»¥ 2: DÃ¹ng `np.concatenate()` vá»›i máº£ng 2D

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

print("GhÃ©p theo hÃ ng (axis=0):\n", np.concatenate((a, b), axis=0))
print("\nGhÃ©p theo cá»™t (axis=1):\n", np.concatenate((a, b), axis=1))
```

### ğŸ“˜ VÃ­ dá»¥ 3: DÃ¹ng `np.stack()` Ä‘á»ƒ thÃªm trá»¥c má»›i

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

stacked = np.stack((arr1, arr2), axis=1)
print(stacked)
```

ğŸ‘‰ **`stack` táº¡o thÃªm má»™t trá»¥c má»›i**, sau Ä‘Ã³ Ä‘áº·t tá»«ng pháº§n tá»­ tÆ°Æ¡ng á»©ng cá»§a arr1 vÃ  arr2 vÃ o hai â€œcá»™tâ€.

- â†’ Táº¡o thÃªm trá»¥c má»›i táº¡i vá»‹ trÃ­ 1
- â†’ GhÃ©p pháº§n tá»­ arr1 vÃ  arr2 láº¡i thÃ nh tá»«ng cáº·p theo chiá»u dá»c
- â†’ Káº¿t quáº£ lÃ  ma tráº­n 3Ã—2.

### ğŸ“˜ VÃ­ dá»¥ 4: GhÃ©p ngang vÃ  dá»c báº±ng `hstack()` vÃ  `vstack()`

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("GhÃ©p ngang:", np.hstack((a, b)))
print("GhÃ©p dá»c:\n", np.vstack((a, b)))
```

## âœ‚ï¸ 3. Splitting Arrays (TÃ¡ch máº£ng)

### ğŸ§  Äá»‹nh nghÄ©a:
TÃ¡ch máº£ng lÃ  **chia má»™t máº£ng lá»›n thÃ nh nhiá»u máº£ng nhá»**.

### âš™ï¸ CÃ¡c hÃ m phá»• biáº¿n:
- `np.split(array, n)` â€” tÃ¡ch Ä‘á»u máº£ng thÃ nh *n pháº§n*.
- `np.array_split(array, n)` â€” tÃ¡ch khÃ´ng Ä‘á»u náº¿u khÃ´ng chia háº¿t.
- `np.hsplit()` â€” tÃ¡ch theo chiá»u ngang (cá»™t).
- `np.vsplit()` â€” tÃ¡ch theo chiá»u dá»c (hÃ ng).

---
### ğŸ“˜ VÃ­ dá»¥ 1: DÃ¹ng `np.split()` Ä‘á»ƒ chia Ä‘á»u máº£ng 1D

```python
arr = np.array([1, 2, 3, 4, 5, 6])

x = np.split(arr, 3)
print(x)
```

### ğŸ“˜ VÃ­ dá»¥ 2: DÃ¹ng `np.array_split()` Ä‘á»ƒ tÃ¡ch khÃ´ng Ä‘á»u

```python
arr = np.array([1, 2, 3, 4, 5, 6, 7])
x = np.array_split(arr, 3)
print(x)
```

### ğŸ“˜ VÃ­ dá»¥ 3: DÃ¹ng `np.hsplit()` vÃ  `np.vsplit()` vá»›i máº£ng 2D

```python
arr2d = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])

print("TÃ¡ch theo cá»™t (hsplit):", np.hsplit(arr2d, 2))
print("\nTÃ¡ch theo hÃ ng (vsplit):", np.vsplit(arr2d, 2))
```
