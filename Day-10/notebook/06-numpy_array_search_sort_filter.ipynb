## ğŸ” 1. Searching Arrays

### ğŸ“˜ VÃ­ dá»¥ 1: TÃ¬m pháº§n tá»­ theo Ä‘iá»u kiá»‡n vá»›i `np.where()`

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
x = np.where(arr > 25)

print("Chá»‰ sá»‘ cÃ¡c pháº§n tá»­ > 25:", x)
print("GiÃ¡ trá»‹ tÆ°Æ¡ng á»©ng:", arr[x])
```

## ğŸ§® 2. Sorting Arrays

Sáº¯p xáº¿p (sorting) lÃ  quÃ¡ trÃ¬nh **thay Ä‘á»•i vá»‹ trÃ­ pháº§n tá»­** Ä‘á»ƒ chÃºng Ä‘Æ°á»£c xáº¿p theo thá»© tá»± tÄƒng dáº§n (hoáº·c giáº£m dáº§n).

## ğŸ¯ 3. Filtering Arrays

### ğŸ§  Äá»‹nh nghÄ©a:
Filtering lÃ  **chá»n cÃ¡c pháº§n tá»­ thá»a Ä‘iá»u kiá»‡n** báº±ng cÃ¡ch táº¡o ra **má»™t máº£ng boolean** (True/False).

Sau Ä‘Ã³, dÃ¹ng máº£ng nÃ y Ä‘á»ƒ trÃ­ch xuáº¥t pháº§n tá»­ tÆ°Æ¡ng á»©ng.

---
### ğŸ“˜ VÃ­ dá»¥ 1: Lá»c pháº§n tá»­ lá»›n hÆ¡n 10

```python
arr = np.array([5, 10, 15, 20, 25])
filter_mask = arr > 10

print("Máº£ng Ä‘iá»u kiá»‡n:", filter_mask)
print("Pháº§n tá»­ thá»a Ä‘iá»u kiá»‡n:", arr[filter_mask])
```

### ğŸ“˜ VÃ­ dá»¥ 2: Káº¿t há»£p nhiá»u Ä‘iá»u kiá»‡n lá»c

```python
arr = np.array([5, 12, 18, 24, 30])

filter_mask = (arr > 10) & (arr < 25)
print("CÃ¡c pháº§n tá»­ tá»« 10 Ä‘áº¿n 25:", arr[filter_mask])
```

### ğŸ“˜ VÃ­ dá»¥ 3: Lá»c giÃ¡ trá»‹ cháºµn/láº»

```python
arr = np.array([1, 2, 3, 4, 5, 6])

even_filter = arr % 2 == 0
odd_filter = arr % 2 != 0

print("Sá»‘ cháºµn:", arr[even_filter])
print("Sá»‘ láº»:", arr[odd_filter])
```
