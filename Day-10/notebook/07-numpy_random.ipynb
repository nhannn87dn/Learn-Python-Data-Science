## âš™ï¸ 1. Táº¡o sá»‘ ngáº«u nhiÃªn cÆ¡ báº£n

Sá»­ dá»¥ng hÃ m `numpy.random.rand()` hoáº·c `numpy.random.randint()` Ä‘á»ƒ sinh sá»‘ ngáº«u nhiÃªn dáº¡ng **float** hoáº·c **int**.

```python
import numpy as np

# Sinh sá»‘ thá»±c ngáº«u nhiÃªn trong khoáº£ng [0, 1)
a = np.random.rand()
print("Sá»‘ thá»±c ngáº«u nhiÃªn:", a)
```

## ğŸ” 3. Chá»n pháº§n tá»­ ngáº«u nhiÃªn (Random Choice)

HÃ m `np.random.choice()` cho phÃ©p chá»n **ngáº«u nhiÃªn pháº§n tá»­** tá»« máº£ng hoáº·c dÃ£y sá»‘, cÃ³ thá»ƒ chá»n 1 hoáº·c nhiá»u pháº§n tá»­.

```python
arr = np.array([10, 20, 30, 40, 50])

# Chá»n 1 pháº§n tá»­ ngáº«u nhiÃªn
print("Má»™t pháº§n tá»­ ngáº«u nhiÃªn:", np.random.choice(arr))
```

```python
# Chá»n 3 pháº§n tá»­ ngáº«u nhiÃªn (cÃ³ thá»ƒ trÃ¹ng)
print("\nBa pháº§n tá»­ ngáº«u nhiÃªn:", np.random.choice(arr, 3))
```

## ğŸ§© 4. Shuffle vÃ  Permutation

Hai phÆ°Æ¡ng thá»©c quan trá»ng Ä‘á»ƒ **trá»™n ngáº«u nhiÃªn (shuffle)** dá»¯ liá»‡u:

- `shuffle()` : trá»™n máº£ng **táº¡i chá»— (in-place)**
- `permutation()` : tráº£ vá» **báº£n sao Ä‘Ã£ Ä‘Æ°á»£c trá»™n**

```python
arr2 = np.array([1, 2, 3, 4, 5])
np.random.shuffle(arr2)
print("Máº£ng sau khi shuffle:", arr2)
```

```python
# permutation táº¡o máº£ng má»›i
arr3 = np.random.permutation([10, 20, 30, 40, 50])
print("\nMáº£ng sau permutation:", arr3)
```
