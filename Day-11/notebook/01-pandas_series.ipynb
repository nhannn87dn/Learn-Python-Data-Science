- L√† **m·ªôt m·∫£ng gi√° tr·ªã (values)** c√≥ **nh√£n (index)** t∆∞∆°ng ·ª©ng.
- H·ªó tr·ª£ **ki·ªÉu d·ªØ li·ªáu h·ªón h·ª£p** (numbers, strings, boolean, objects‚Ä¶)

### ‚öôÔ∏è 2.2. C√°ch t·∫°o Series
#### üîπ T·∫°o t·ª´ `list`

```python
s = pd.Series([10, 20, 30, 40])
print(s)
```

#### üîπ T·∫°o v·ªõi index t√πy ch·ªânh

```python
s = pd.Series([10, 20, 30], index=['A', 'B', 'C'])
print(s)
```

#### üîπ T·∫°o t·ª´ dictionary

```python
data = {'a': 100, 'b': 200, 'c': 300}
s = pd.Series(data)
print(s)
```

#### üîπ T·∫°o t·ª´ NumPy array

```python
arr = np.array([1, 2, 3, 4])
s = pd.Series(arr, index=['A', 'B', 'C', 'D'])
print(s)
```

#### üîπ T·∫°o Series v·ªõi gi√° tr·ªã h·∫±ng

```python
s = pd.Series(7, index=['a', 'b', 'c'])
print(s)
```

### üßæ 2.3. Thu·ªôc t√≠nh c·ªßa Series

```python
# In ra m·∫£ng c√°c gi√° tr·ªã (gi·ªëng nh∆∞ NumPy array)
print('values --> ', s.values)
# In ra nh√£n (index) c·ªßa Series
print('index --> ',s.index)
# In ra ki·ªÉu d·ªØ li·ªáu c·ªßa c√°c ph·∫ßn t·ª≠ trong Series
print('dtype --> ',s.dtype)
# In ra k√≠ch th∆∞·ªõc h√¨nh d·∫°ng c·ªßa Series (s·ªë ph·∫ßn t·ª≠)
print('shape --> ',s.shape)
# In ra t·ªïng s·ªë ph·∫ßn t·ª≠ trong Series
print('size --> ',s.size)
```

### üîç 2.4. Truy c·∫≠p d·ªØ li·ªáu trong Series
#### üîπ Theo v·ªã tr√≠ (gi·ªëng list/array)

```python
s = pd.Series([10, 20, 30, 40], index=['A', 'B', 'C', 'D'])
# D·ª±a v√†o index
print(s[0])
```

```python
# C√∫ ph√°p series[start:stop]
print(s[:2])
```

#### üîπ Theo nh√£n - label (index)

```python
print(s['B']) # M·ªôt v·ªã tr√≠ c·ª• th·ªÉ
```

```python
print(s[['A', 'D']]) # Ho·∫∑c nhi·ªÅu
```

#### üîπ D√πng `.loc[]` v√† `.iloc[]`
C√°ch d√πng n√†y ƒë∆∞·ª£c khuy·∫øn ngh·ªã h∆°n

```python
print(s.loc['C']) # D·ª±a v√†o location c·ªßa label c·ª• th·ªÉ
```

```python
print(s.iloc[2]) # D·ª±a v√†o index location c·ªßa ph·∫ßn t·ª≠
```

### üßÆ 2.5. To√°n t·ª≠ & thao t√°c tr√™n Series

```python
a = pd.Series([10, 20, 30])
b = pd.Series([1, 2, 3])
print('a + b => ', a + b)
print('a * b => ',a * 2)
print('a / b => ',a / b)
```

#### ‚ö†Ô∏è N·∫øu index kh√¥ng tr√πng, Pandas s·∫Ω t·ª± cƒÉn ch·ªânh theo index

```python
a = pd.Series([10, 20, 30], index=['x', 'y', 'z'])
b = pd.Series([1, 2, 3], index=['y', 'z', 'w'])
print(a + b)
```

### üî¢ 2.6. M·ªôt s·ªë ph∆∞∆°ng th·ª©c ph·ªï bi·∫øn

| Ph∆∞∆°ng th·ª©c      | M√¥ t·∫£                        |
| ---------------- | ---------------------------- |
| `head(n)`        | L·∫•y n ph·∫ßn t·ª≠ ƒë·∫ßu            |
| `tail(n)`        | L·∫•y n ph·∫ßn t·ª≠ cu·ªëi           |
| `sum()`          | T·ªïng gi√° tr·ªã                 |
| `mean()`         | Trung b√¨nh                   |
| `max() / min()`  | Gi√° tr·ªã l·ªõn/nh·ªè nh·∫•t         |
| `sort_values()`  | S·∫Øp x·∫øp theo gi√° tr·ªã         |
| `sort_index()`   | S·∫Øp x·∫øp theo index           |
| `unique()`       | L·∫•y gi√° tr·ªã duy nh·∫•t         |
| `value_counts()` | ƒê·∫øm t·∫ßn su·∫•t gi√° tr·ªã         |
| `apply(func)`    | √Åp d·ª•ng h√†m cho t·ª´ng ph·∫ßn t·ª≠ |


```python
# T·∫°o Series v√≠ d·ª•
s = pd.Series([10, 20, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e', 'f'])
```

üß© head(n) ‚Äì L·∫•y n ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n

```python
# L·∫•y 3 ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n
print(s.head(3))
```

üß© tail(n) ‚Äì L·∫•y n ph·∫ßn t·ª≠ cu·ªëi c√πng

```python
# L·∫•y 2 ph·∫ßn t·ª≠ cu·ªëi c√πng
print(s.tail(2))
```

üß© sum() ‚Äì T√≠nh t·ªïng c√°c gi√° tr·ªã

```python
# T√≠nh gi√° tr·ªã trung b√¨nh
print("Trung b√¨nh =", s.mean())
```

üß© max() v√† min() ‚Äì Gi√° tr·ªã l·ªõn nh·∫•t v√† nh·ªè nh·∫•t

```python
print("Gi√° tr·ªã l·ªõn nh·∫•t =", s.max())
print("Gi√° tr·ªã nh·ªè nh·∫•t =", s.min())
```

üß© sort_values() ‚Äì S·∫Øp x·∫øp theo gi√° tr·ªã

```python
# S·∫Øp x·∫øp tƒÉng d·∫ßn
print("S·∫Øp x·∫øp tƒÉng d·∫ßn:\n", s.sort_values())

# S·∫Øp x·∫øp gi·∫£m d·∫ßn
print("\nS·∫Øp x·∫øp gi·∫£m d·∫ßn:\n", s.sort_values(ascending=False))
```

üß© sort_index() ‚Äì S·∫Øp x·∫øp theo ch·ªâ m·ª•c (index)

```python
# S·∫Øp x·∫øp theo t√™n ch·ªâ m·ª•c
print("S·∫Øp x·∫øp theo index:\n", s.sort_index())
```

üß© unique() ‚Äì L·∫•y ra c√°c gi√° tr·ªã duy nh·∫•t

```python
print("C√°c gi√° tr·ªã duy nh·∫•t:", s.unique())
```

üß© value_counts() ‚Äì ƒê·∫øm t·∫ßn su·∫•t xu·∫•t hi·ªán

```python
print("ƒê·∫øm t·∫ßn su·∫•t xu·∫•t hi·ªán:\n", s.value_counts())
```

üß© apply(func) ‚Äì √Åp d·ª•ng h√†m l√™n t·ª´ng ph·∫ßn t·ª≠

```python
# Nh√¢n m·ªói gi√° tr·ªã v·ªõi 10
print("Nh√¢n m·ªói ph·∫ßn t·ª≠ v·ªõi 10:\n", s.apply(lambda x: x * 10))

# H√†m t√πy ch·ªânh: ph√¢n lo·∫°i gi√° tr·ªã
def category(val):
    if val < 20:
        return "Th·∫•p"
    elif val < 40:
        return "Trung b√¨nh"
    else:
        return "Cao"

print("\nPh√¢n lo·∫°i gi√° tr·ªã:\n", s.apply(category))
```

### üîÑ 2.7. Ki·ªÉm tra & x·ª≠ l√Ω d·ªØ li·ªáu thi·∫øu (NaN)

```python
s = pd.Series([1, 2, None, 4])

print(s.isnull())      # [False False True False]
print(s.notnull())     # [ True  True False  True]
print(s.fillna(0))     # ƒêi·ªÅn gi√° tr·ªã 0 v√†o ch·ªó NaN
print(s.dropna())      # Lo·∫°i b·ªè gi√° tr·ªã NaN
```

### üßæ 2.8. K·∫øt h·ª£p Series

B·∫°n c√≥ th·ªÉ **c·ªông, gh√©p n·ªëi (concat), ho·∫∑c bi·∫øn ƒë·ªïi** nhi·ªÅu Series v·ªõi nhau.

```python
s1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s2 = pd.Series([4, 5, 6], index=['d', 'e', 'f'])
s3 = pd.concat([s1, s2])
print(s3)
```
