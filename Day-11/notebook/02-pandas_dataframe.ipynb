## üß© 3.2. C√°ch t·∫°o DataFrame
### üîπ T·∫°o t·ª´ dictionary

```python
import pandas as pd

data = {
    'Name': ['An', 'B√¨nh', 'Chi'],
    'Age': [20, 21, 19],
    'Score': [85, 90, 88]
}

df = pd.DataFrame(data)
print(df)
```

### üîπ T·∫°o t·ª´ danh s√°ch c√°c dictionary

```python
data = [
    {'Name': 'An', 'Age': 20, 'Score': 85},
    {'Name': 'B√¨nh', 'Age': 21, 'Score': 90},
    {'Name': 'Chi', 'Age': 19, 'Score': 88}
]

df = pd.DataFrame(data)
print(df)
```

### üîπ T·∫°o t·ª´ NumPy array

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
df = pd.DataFrame(arr, columns=['A', 'B', 'C'])
print(df)
```

### üîπ T·∫°o DataFrame r·ªóng v√† th√™m c·ªôt sau

```python
df = pd.DataFrame()
df['Name'] = ['An', 'B√¨nh']
df['Score'] = [85, 92]
print(df)
```

## üîç 3.3. Truy c·∫≠p d·ªØ li·ªáu

### üß© Truy c·∫≠p c·ªôt

```python
print(df['Name'])
print(df.Score)
```

### üß© Truy c·∫≠p h√†ng b·∫±ng `loc` v√† `iloc`

```python
# Truy c·∫≠p h√†ng theo index label
print(df.loc[0])  # H√†ng ƒë·∫ßu ti√™n
```

```python
df.loc[0, 'Name']  # l·∫•y gi√° tr·ªã c·ª• th·ªÉ
```

```python
# Truy c·∫≠p h√†ng theo v·ªã tr√≠ s·ªë
print(df.iloc[1])  # H√†ng th·ª© 2
```

```python
df.iloc[1:3] # T·ª´ h√†ng 1 ƒë·∫øn 2
```

### üß© Truy c·∫≠p v√πng d·ªØ li·ªáu con (slice)

```python
print(df.loc[0:1, ['Name', 'Score']])
```

## ‚úèÔ∏è 3.4. Th√™m v√† x√≥a d·ªØ li·ªáu

## üßæ 3.6. L·ªçc d·ªØ li·ªáu theo ƒëi·ªÅu ki·ªán

```python
print(df[df['Score'] > 85])  # L·ªçc theo ƒëi·ªÉm
print(df[(df['Age'] > 19) & (df['Score'] >= 85)])  # K·∫øt h·ª£p nhi·ªÅu ƒëi·ªÅu ki·ªán
```

## üîó 3.7. G·ªôp & n·ªëi DataFrame

### üîπ `concat()` ‚Äì Gh√©p d·ªçc ho·∫∑c ngang

```python
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

print(pd.concat([df1, df2], axis=0))  # Gh√©p d·ªçc (th√™m h√†ng)
print(pd.concat([df1, df2], axis=1))  # Gh√©p ngang (th√™m c·ªôt)
```

### üîπ `merge()` ‚Äì G·ªôp theo c·ªôt chung (gi·ªëng JOIN trong SQL)

```python
df_left = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['An', 'B√¨nh', 'Chi']})
df_right = pd.DataFrame({'ID': [1, 2, 4], 'Score': [85, 90, 70]})

merged = pd.merge(df_left, df_right, on='ID', how='inner')
print(merged)
```

### üîπ M·ªôt s·ªë thao t√°c h·ªØu √≠ch kh√°c

```python
df['Age'].mean()       # Trung b√¨nh tu·ªïi
df['Name'].unique()    # C√°c gi√° tr·ªã duy nh·∫•t
df['Score'].value_counts() # ƒê·∫øm t·∫ßn su·∫•t
df.rename(columns={'Name': 'FullName'}, inplace=True) # ƒë·ªïi t√™n c·ªôt
```

### üîπ  Truy xu·∫•t n√¢ng cao
#### √Åp d·ª•ng h√†m cho t·ª´ng ph·∫ßn t·ª≠

```python
# C·ªông v√†o tr∆∞·ªùng Age th√™m 1 ƒë∆°n v·ªã
df['Age'] = df['Age'].apply(lambda x: x + 1)
```

#### G·ªôp nh√≥m (Group By)

```python
#T√≠nh trung b√¨nh c·ªôt Age
df.groupby('City')['Age'].mean()
```

## üìä **4.9. K·ªπ thu·∫≠t di chuy·ªÉn v·ªã tr√≠ c√°c c·ªôt trong DataFrame**

Gi·∫£ s·ª≠ b·∫°n c√≥ DataFrame nh∆∞ sau:

```python
df3 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 22],
    'City': ['New York', 'London', 'Tokyo']
})

print(df3)
```

**1. C√°ch 1 ‚Äì S·∫Øp x·∫øp l·∫°i b·∫±ng danh s√°ch t√™n c·ªôt**

N·∫øu b·∫°n mu·ªën **di chuy·ªÉn c·ªôt `City` l√™n ƒë·∫ßu**, b·∫°n c√≥ th·ªÉ ch·ªâ ƒë·ªãnh th·ª© t·ª± c·ªôt th·ªß c√¥ng:

```python
c_1 = df3[['City', 'Name', 'Age']]
print(c_1)
```

C√°ch 2 ‚Äì D√πng `insert()` ƒë·ªÉ ch√®n l·∫°i c·ªôt v√†o v·ªã tr√≠ mong mu·ªën**

B·∫°n c√≥ th·ªÉ **ch√®n m·ªôt c·ªôt ƒë√£ c√≥ v√†o v·ªã tr√≠ m·ªõi** r·ªìi x√≥a c·ªôt c≈©.

V√≠ d·ª•: di chuy·ªÉn c·ªôt `City` l√™n v·ªã tr√≠ ƒë·∫ßu ti√™n:

```python
city_col = df3.pop('City')      # L·∫•y v√† x√≥a c·ªôt City kh·ªèi DataFrame
df3.insert(0, 'City', city_col) # Ch√®n l·∫°i c·ªôt City v√†o v·ªã tr√≠ ƒë·∫ßu (index 0)
print(df3)
```

üëâ ∆Øu ƒëi·ªÉm:

* D·ªÖ d√πng khi ch·ªâ mu·ªën di chuy·ªÉn **m·ªôt c·ªôt duy nh·∫•t**.
* Kh√¥ng c·∫ßn vi·∫øt l·∫°i to√†n b·ªô danh s√°ch c·ªôt.

C√°ch 3 ‚Äì S·ª≠ d·ª•ng `reindex()` ƒë·ªÉ s·∫Øp x·∫øp l·∫°i c·ªôt**

C√°ch n√†y h·ªØu √≠ch khi b·∫°n mu·ªën ƒë·∫£m b·∫£o DataFrame c√≥ **th·ª© t·ª± c·ªôt c·ªë ƒë·ªãnh**, c√≥ th·ªÉ √°p d·ª•ng sau khi merge ho·∫∑c ƒë·ªçc t·ª´ file.

```python
df3 = df3.reindex(columns=['Age', 'Name', 'City'])
print(df3)
```
