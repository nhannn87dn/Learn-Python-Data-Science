## 3. C√°ch v·∫ª bi·ªÉu ƒë·ªì v·ªõi seaborn

### 3.1 Bi·ªÉu ƒë·ªì Line

````python
# T·∫°o m·ªôt DataFrame m·∫´u
data = {
    'NƒÉm': [2018, 2019, 2020, 2021, 2022],
    'Doanh thu': [200, 300, 400, 500, 600]
}
df = pd.DataFrame(data)

# Seaborn - V·∫Ω bi·ªÉu ƒë·ªì ƒë∆∞·ªùng
sns.lineplot(x='NƒÉm', y='Doanh thu', data=df)

# Th√™m ti√™u ƒë·ªÅ v√† nh√£n
plt.title('Bi·ªÉu ƒë·ªì Doanh thu theo NƒÉm')
plt.xlabel('NƒÉm')
plt.ylabel('Doanh thu')
# B·∫≠t l∆∞·ªõi
plt.grid(True) 

# Matplotlib - Hi·ªÉn th·ªã bi·ªÉu ƒë·ªì
plt.show()
````

### 3.2. Bi·ªÉu ƒë·ªì ph√¢n t√°n (Scatter plot)

Th·ªÉ hi·ªán m·ªëi quan h·ªá gi·ªØa 2 bi·∫øn s·ªë (gi·ªëng nh∆∞ bi·ªÉu ƒë·ªì scatter c·ªßa Matplotlib, nh∆∞ng d·ªÖ t√πy ch·ªânh h∆°n).

````python
# Load m·ªôt DataFrame m·∫´u t·ª´ seaborn
df = sns.load_dataset("tips")
df.head()

# V·∫Ω bi·ªÉu ƒë·ªì ph√¢n t√°n
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", style="smoker")
plt.title("M·ªëi quan h·ªá gi·ªØa ho√° ƒë∆°n v√† ti·ªÅn tip")


# Th√™m ti√™u ƒë·ªÅ v√† nh√£n
plt.xlabel('Total Bill')
plt.ylabel('Tip')

# Hi·ªÉn th·ªã bi·ªÉu ƒë·ªì
plt.show()
````

**Gi·∫£i th√≠ch:**

* `hue`: t√¥ m√†u theo gi·ªõi t√≠nh.
* `style`: k√Ω hi·ªáu ƒëi·ªÉm theo nh√≥m.
* `size`: k√≠ch th∆∞·ªõc ƒëi·ªÉm theo gi√° tr·ªã kh√°c.

### 3.3  Bi·ªÉu ƒë·ªì c·ªôt (Bar plot)

Hi·ªÉn th·ªã gi√° tr·ªã trung b√¨nh c·ªßa m·ªôt bi·∫øn theo nh√≥m.

````python
# T·∫°o m·ªôt DataFrame m·∫´u
data = {
    'NƒÉm': [2018, 2019, 2020, 2021, 2022],
    'Doanh thu': [200, 300, 400, 500, 600]
}
df = pd.DataFrame(data)

# V·∫Ω bi·ªÉu ƒë·ªì c·ªôt
sns.barplot(x='NƒÉm', y='Doanh thu', data=df)

# Th√™m ti√™u ƒë·ªÅ v√† nh√£n
plt.title('Bi·ªÉu ƒë·ªì C·ªôt: Doanh thu theo NƒÉm')
plt.xlabel('NƒÉm')
plt.ylabel('Doanh thu')

# Hi·ªÉn th·ªã bi·ªÉu ƒë·ªì
plt.show()
````

**Gi·∫£i th√≠ch:**

* `ci=None`: t·∫Øt hi·ªÉn th·ªã kho·∫£ng tin c·∫≠y (confidence interval).

### 3.4  Bi·ªÉu ƒë·ªì ma tr·∫≠n t∆∞∆°ng quan 

Hi·ªÉn th·ªã m·ªëi t∆∞∆°ng quan gi·ªØa c√°c bi·∫øn s·ªë.

````python

import numpy as np

# Gi·∫£ l·∫≠p d·ªØ li·ªáu t·∫ßn su·∫•t th√¥ng quan container theo gi·ªù trong su·ªët m·ªôt tu·∫ßn
data = np.random.randint(0, 100, size=(24, 7))  # T·∫°o m·∫£ng ng·∫´u nhi√™n c√≥ k√≠ch th∆∞·ªõc 24x7
dates = ['Th·ª© Hai', 'Th·ª© Ba', 'Th·ª© T∆∞', 'Th·ª© NƒÉm', 'Th·ª© S√°u', 'Th·ª© B·∫£y', 'Ch·ªß Nh·∫≠t']
hours = [f'{i}:00' for i in range(24)]

# T·∫°o DataFrame
heatmap_data = pd.DataFrame(data, columns=dates, index=hours)

print(heatmap_data)


# V·∫Ω bi·ªÉu ƒë·ªì nhi·ªát
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap='YlGnBu')

# Th√™m ti√™u ƒë·ªÅ
plt.title('T·∫ßn su·∫•t Container Th√¥ng quan theo Gi·ªù trong Tu·∫ßn')
plt.xlabel('Ng√†y trong Tu·∫ßn')
plt.ylabel('Gi·ªù trong Ng√†y')

# Hi·ªÉn th·ªã bi·ªÉu ƒë·ªì
plt.show()
````
