## 3. S·ª≠ d·ª•ng Matplotib c∆° b·∫£n

### 3.1 Matplotlib Plotting

#### 3.1.1 V·∫ª 2 ƒëi·ªÉm d·ª±a tr√™n to·∫° ƒë·ªô `x` v√† `y`

H√†m **plot()** ƒë∆∞·ª£c s·ª≠ d·ª•ng ƒë·ªÉ **v·∫Ω c√°c ƒëi·ªÉm (markers) tr√™n bi·ªÉu ƒë·ªì**.

Theo m·∫∑c ƒë·ªãnh, **h√†m plot() v·∫Ω m·ªôt ƒë∆∞·ªùng n·ªëi t·ª´ ƒëi·ªÉm n√†y ƒë·∫øn ƒëi·ªÉm kh√°c**.

H√†m n√†y nh·∫≠n c√°c tham s·ªë ƒë·ªÉ **x√°c ƒë·ªãnh c√°c ƒëi·ªÉm tr√™n bi·ªÉu ƒë·ªì**:

* **Tham s·ªë 1** l√† m·ªôt m·∫£ng ch·ª©a c√°c ƒëi·ªÉm tr√™n **tr·ª•c x**.
* **Tham s·ªë 2** l√† m·ªôt m·∫£ng ch·ª©a c√°c ƒëi·ªÉm tr√™n **tr·ª•c y**.

V√≠ d·ª•: n·∫øu mu·ªën v·∫Ω m·ªôt ƒë∆∞·ªùng t·ª´ ƒëi·ªÉm **(1, 3)** ƒë·∫øn **(8, 10)**, ch√∫ng ta c·∫ßn truy·ªÅn hai m·∫£ng `[1, 8]` v√† `[3, 10]` v√†o h√†m **plot()**.


````python
import numpy as np
# Ch·∫°y ƒë·ªÉ xem k·∫øt qu·∫£
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
````

#### 3.1.2 V·∫ª NHI·ªÄU ƒëi·ªÉm d·ª±a tr√™n to·∫° ƒë·ªô `x` v√† `y`

B·∫°n s·∫Ω c·∫ßn cung c·∫•p ƒë·∫ßu v√†o nhi·ªÅu h∆°n nh∆∞ sau:

````python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
````

#### 3.1.3 M·∫∑c ƒë·ªãnh c√°c ƒëi·ªÉm c·ªßa x

N·∫øu b·∫°n kh√¥ng cung c·∫•p ƒë·∫ßu v√†o cho tr·ª•c `x` th√¨ n√≥ s·∫Ω t·ª± sinh ra d·ª±a tr√™n s·ªë l∆∞·ª£ng ph·∫ßn t·ª´ c·ªßa tr·ª•c `y`

````python
#Ch·ªâ cung c·∫•p y-points
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
#Ch·∫°y ƒë·ªÉ xem k·∫øt qu·∫£
````

`x-points` trong v√≠ d·ª• tr√™n l√† [0, 1, 2, 3, 4, 5].

#### 3.1.2 V·∫ª c√°c ƒëi·ªÉm nh∆∞ng kh√¥ng n·ªëi nhau

Gi·ªëng nh∆∞ ƒë√°nh d·∫•u c√°c ƒëi·ªÉm d·ª±a v√†o to·∫° ƒë·ªô `x` v√† `y` c·ªßa ch√∫ng tr√™n bi·ªÉu ƒë·ªì.

````python
# Ch·∫°y ƒë·ªÉ xem k·∫øt qu·∫£
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
````

### 3.2 Matplotlib Maker

**Marker** l√† k√Ω hi·ªáu ƒë·∫°i di·ªán cho m·ªói ƒëi·ªÉm d·ªØ li·ªáu tr√™n bi·ªÉu ƒë·ªì.

V√≠ d·ª•: d·∫•u ch·∫•m, h√¨nh tr√≤n, d·∫•u c·ªông, d·∫•u sao‚Ä¶

B·∫°n c√≥ th·ªÉ s·ª≠ d·ª•ng ƒë·ªïi s·ªë `marker` ƒë·ªÉ nh·∫•n m·∫°nh t·ª´ng ƒëi·ªÉm v·ªõi m·ªôt d·∫•u hi·ªáu ƒë∆∞·ª£c ch·ªâ ƒë·ªãnh

V√≠ d·ª•:

````python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
# Ho·∫∑c
# plt.plot(ypoints, marker = '*')

plt.show()
#Ch·∫°y ƒë·ªÉ xem
````

**M·ªôt s·ªë ki·ªÉu marker ph·ªï bi·∫øn:**

| Marker | M√¥ t·∫£                      |
| ------ | -------------------------- |
| `'.'`  | ƒêi·ªÉm ch·∫•m nh·ªè              |
| `','`  | Pixel                      |
| `'o'`  | H√¨nh tr√≤n                  |
| `'s'`  | H√¨nh vu√¥ng (square)        |
| `'D'`  | H√¨nh thoi (diamond)        |
| `'+'`  | D·∫•u c·ªông                   |
| `'x'`  | D·∫•u x                      |
| `'*'`  | Sao                        |
| `'p'`  | Ng√¥i sao 5 c√°nh (pentagon) |
| `'h'`  | L·ª•c gi√°c (hexagon1)        |


### 3.2 Matplotlib Maker Size

* Tham s·ªë: `markersize` ho·∫∑c `ms` --> K√≠ch th∆∞·ªõc c·ªßa marker
* M·∫∑c ƒë·ªãnh: `6`
* TƒÉng gi√° tr·ªã ‚Üí marker l·ªõn h∆°n, gi·∫£m gi√° tr·ªã ‚Üí nh·ªè h∆°n.


````python
#V√≠ d·ª•
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y, marker='o', markersize=10)  # marker h√¨nh tr√≤n, size 10
plt.show()
````

### 3.3 Matplotlib Maker Color

* Tham s·ªë: `markerfacecolor` (`mfc`) ho·∫∑c `markeredgecolor` (`mec`)
* `markerfacecolor` ‚Üí m√†u b√™n trong marker
* `markeredgecolor` ‚Üí m√†u vi·ªÅn marker

````python
#V√≠ d·ª•
plt.plot(x, y, marker='s', markersize=12, markerfacecolor='red', markeredgecolor='blue')
plt.show()
````

* Trong v√≠ d·ª• tr√™n: marker l√† h√¨nh vu√¥ng, m√†u ƒë·ªè b√™n trong, vi·ªÅn xanh.

* B·∫°n c√≥ th·ªÉ v·∫Ω **ƒë∆∞·ªùng n·ªëi c√°c ƒëi·ªÉm** v√† ƒë√°nh d·∫•u c√°c ƒëi·ªÉm b·∫±ng marker c√πng l√∫c.


````python
plt.plot(x, y, linestyle='--', marker='D', markersize=8, markerfacecolor='yellow', markeredgecolor='green', color='purple')
plt.show()
````

* `linestyle='--'` ‚Üí ƒë∆∞·ªùng n√©t ƒë·ª©t
* `color='purple'` ‚Üí m√†u ƒë∆∞·ªùng

### 3.4 Matplotlib Line

Trong **Matplotlib**, khi v·∫Ω bi·ªÉu ƒë·ªì b·∫±ng `plot()`, b·∫°n c√≥ th·ªÉ **t√πy ch·ªânh ƒë∆∞·ªùng n·ªëi gi·ªØa c√°c ƒëi·ªÉm d·ªØ li·ªáu**. ƒê∆∞·ªùng n√†y gi√∫p bi·ªÉu di·ªÖn **xu h∆∞·ªõng d·ªØ li·ªáu**, v√† c√≥ th·ªÉ **t√πy bi·∫øn v·ªÅ ki·ªÉu, m√†u, v√† ƒë·ªô d√†y**.

**1. Line Style (Ki·ªÉu ƒë∆∞·ªùng)**

* Tham s·ªë: `linestyle` ho·∫∑c `ls`
* M·ªôt s·ªë ki·ªÉu ph·ªï bi·∫øn:

| Ki·ªÉu                    | M√¥ t·∫£                 |
| ----------------------- | --------------------- |
| `'-'`                   | ƒê∆∞·ªùng li·ªÅn (m·∫∑c ƒë·ªãnh) |
| `'--'`                  | ƒê∆∞·ªùng n√©t ƒë·ª©t         |
| `'-.'`                  | ƒê∆∞·ªùng ch·∫•m g·∫°ch       |
| `':'`                   | ƒê∆∞·ªùng ch·∫•m ch·∫•m       |
| `'None'` / `' '` / `''` | Kh√¥ng v·∫Ω ƒë∆∞·ªùng        |


````python
# V√≠ d·ª•
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y, linestyle='--')  # ƒë∆∞·ªùng n√©t ƒë·ª©t
plt.show()
````

**2. Line Width (ƒê·ªô d√†y ƒë∆∞·ªùng)**

* Tham s·ªë: `linewidth` ho·∫∑c `lw`
* M·∫∑c ƒë·ªãnh: `1.5`
* Gi√° tr·ªã l·ªõn ‚Üí ƒë∆∞·ªùng d√†y h∆°n, nh·ªè ‚Üí m·ªèng h∆°n.

````python
plt.plot(x, y, linestyle='-', linewidth=4)
plt.show()
````

**3. Line Color (M√†u ƒë∆∞·ªùng)**

* Tham s·ªë: `color` ho·∫∑c `c`
* C√≥ th·ªÉ d√πng: t√™n m√†u (`'red'`), m√£ HEX (`'#FF5733'`), k√Ω hi·ªáu m√†u (`'b','g','r','c','m','y','k','w'`)

````python
plt.plot(x, y, linestyle='-', color='green', linewidth=3)
plt.show()
````

**4. K·∫øt h·ª£p Line v·ªõi Marker**

* B·∫°n c√≥ th·ªÉ **v·∫Ω ƒë∆∞·ªùng n·ªëi c√°c ƒëi·ªÉm** ƒë·ªìng th·ªùi ƒë√°nh d·∫•u c√°c ƒëi·ªÉm b·∫±ng **marker**.

````python
#V√≠ d·ª• t·ªïng h·ª£p
plt.plot(x, y,
         linestyle='-.',       # Ki·ªÉu ƒë∆∞·ªùng
         linewidth=2,          # ƒê·ªô d√†y
         color='purple',       # M√†u ƒë∆∞·ªùng
         marker='o',           # Ki·ªÉu marker
         markersize=8,         # K√≠ch th∆∞·ªõc marker
         markerfacecolor='orange', # M√†u b√™n trong marker
         markeredgecolor='black'   # M√†u vi·ªÅn marker
        )
plt.show()
````

**5. T√≥m t·∫Øt c√°c tham s·ªë Line trong `plot()`**

| Tham s·ªë                   | M√¥ t·∫£                                 |
| ------------------------- | ------------------------------------- |
| `linestyle` / `ls`        | Ki·ªÉu ƒë∆∞·ªùng n·ªëi ('-', '--', '-.', ':') |
| `linewidth` / `lw`        | ƒê·ªô d√†y ƒë∆∞·ªùng n·ªëi                      |
| `color` / `c`             | M√†u ƒë∆∞·ªùng                             |
| `marker`                  | Ki·ªÉu marker (n·∫øu mu·ªën ƒë√°nh d·∫•u ƒëi·ªÉm)  |
| `markersize` / `ms`       | K√≠ch th∆∞·ªõc marker                     |
| `markerfacecolor` / `mfc` | M√†u b√™n trong marker                  |
| `markeredgecolor` / `mec` | M√†u vi·ªÅn marker                       |


### 3.5 Matplotlib Label

Trong **Matplotlib**, khi v·∫Ω bi·ªÉu ƒë·ªì, vi·ªác th√™m **nh√£n tr·ª•c (labels) v√† ti√™u ƒë·ªÅ (title)** gi√∫p **bi·ªÉu ƒë·ªì d·ªÖ ƒë·ªçc v√† chuy√™n nghi·ªáp h∆°n**.


**1. Title (Ti√™u ƒë·ªÅ bi·ªÉu ƒë·ªì)**

* **H√†m**: `plt.title()`
* **M·ª•c ƒë√≠ch**: Hi·ªÉn th·ªã **ti√™u ƒë·ªÅ t·ªïng quan** cho bi·ªÉu ƒë·ªì, gi√∫p ng∆∞·ªùi ƒë·ªçc hi·ªÉu n·ªôi dung bi·ªÉu ƒë·ªì.
* **Tham s·ªë quan tr·ªçng**:

  * `label` (ho·∫∑c tr·ª±c ti·∫øp gi√° tr·ªã ƒë·∫ßu v√†o): n·ªôi dung ti√™u ƒë·ªÅ
  * `fontsize`: k√≠ch th∆∞·ªõc ch·ªØ
  * `color`: m√†u ch·ªØ
  * `loc`: v·ªã tr√≠ (`'left'`, `'center'` (m·∫∑c ƒë·ªãnh), `'right'`)


**V√≠ d·ª•:**

````python
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
plt.title("Doanh thu theo th√°ng", fontsize=16, color='blue', loc='center')
plt.show()
````

**2. Labels (Nh√£n tr·ª•c)**

* **H√†m**:

  * `plt.xlabel()` ‚Üí nh√£n tr·ª•c X
  * `plt.ylabel()` ‚Üí nh√£n tr·ª•c Y
* **M·ª•c ƒë√≠ch**: Gi·∫£i th√≠ch **√Ω nghƒ©a c·ªßa t·ª´ng tr·ª•c**, gi√∫p ng∆∞·ªùi ƒë·ªçc hi·ªÉu d·ªØ li·ªáu hi·ªÉn th·ªã.
* **Tham s·ªë**:

  * `label` (ho·∫∑c gi√° tr·ªã tr·ª±c ti·∫øp): n·ªôi dung nh√£n
  * `fontsize`: k√≠ch th∆∞·ªõc ch·ªØ
  * `color`: m√†u ch·ªØ
  * `rotation`: xoay nh√£n (th∆∞·ªùng d√πng cho tr·ª•c X khi t√™n d√†i)

**V√≠ d·ª•:**

````python
plt.plot(x, y)
plt.xlabel("Th√°ng", fontsize=12, color='green')
plt.ylabel("Doanh thu (tri·ªáu VND)", fontsize=12, color='red')
plt.show()
````

**3. K·∫øt h·ª£p Title v√† Labels**

B·∫°n c√≥ th·ªÉ d√πng ƒë·ªìng th·ªùi **title + xlabel + ylabel** ƒë·ªÉ bi·ªÉu ƒë·ªì ƒë·∫ßy ƒë·ªß th√¥ng tin:


````python
plt.plot(x, y, marker='o', linestyle='--', color='purple')
plt.title("Doanh thu theo th√°ng", fontsize=16, color='blue')
plt.xlabel("Th√°ng", fontsize=12, color='green')
plt.ylabel("Doanh thu (tri·ªáu VND)", fontsize=12, color='red')
plt.grid(True)
plt.show()
````

* `plt.grid(True)` ‚Üí hi·ªÉn th·ªã l∆∞·ªõi ƒë·ªÉ d·ªÖ ƒë·ªçc gi√° tr·ªã.


**L∆∞u √Ω:**

* N√™n lu√¥n ƒë·∫∑t **title + xlabel + ylabel** khi v·∫Ω bi·ªÉu ƒë·ªì ƒë·ªÉ bi·ªÉu ƒë·ªì **tr·ª±c quan, d·ªÖ hi·ªÉu**.
* C√≥ th·ªÉ k·∫øt h·ª£p v·ªõi **marker, line, grid** ƒë·ªÉ bi·ªÉu ƒë·ªì ho√†n ch·ªânh.

### 3.6 Matplotlib Grid

Trong **Matplotlib**, **grid** l√† **l∆∞·ªõi tham chi·∫øu ƒë∆∞·ª£c v·∫Ω ph√≠a sau c√°c ƒë∆∞·ªùng v√† marker c·ªßa bi·ªÉu ƒë·ªì**.

* **M·ª•c ƒë√≠ch**: Gi√∫p **ƒë·ªçc gi√° tr·ªã d·ªÖ d√†ng**, x√°c ƒë·ªãnh kho·∫£ng c√°ch, v√† tƒÉng t√≠nh tr·ª±c quan c·ªßa bi·ªÉu ƒë·ªì.
* Grid th∆∞·ªùng ƒë∆∞·ª£c s·ª≠ d·ª•ng **c√πng v·ªõi line plot, bar chart, scatter plot‚Ä¶**.


**1. B·∫≠t / t·∫Øt Grid**

* **H√†m**: `plt.grid()`
* **Tham s·ªë**:

  * `True` ‚Üí b·∫≠t l∆∞·ªõi
  * `False` ‚Üí t·∫Øt l∆∞·ªõi (m·∫∑c ƒë·ªãnh)

````python
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y, marker='o')
plt.grid(True)  # b·∫≠t l∆∞·ªõi
plt.show()
````

**2. T√πy ch·ªânh Grid**

* **C√°c tham s·ªë quan tr·ªçng**:

  * `color` ‚Üí m√†u l∆∞·ªõi
  * `linestyle` ‚Üí ki·ªÉu ƒë∆∞·ªùng l∆∞·ªõi (`'-'`, `'--'`, `':'`, `'-.')`
  * `linewidth` ‚Üí ƒë·ªô d√†y ƒë∆∞·ªùng l∆∞·ªõi
  * `alpha` ‚Üí ƒë·ªô trong su·ªët (0: trong su·ªët, 1: m·ªù)

**V√≠ d·ª•:**

````python
plt.plot(x, y, marker='s', color='purple')
plt.grid(True, color='gray', linestyle='--', linewidth=0.8, alpha=0.5)
plt.show()
````

**3. Grid theo tr·ª•c**

* B·∫°n c√≥ th·ªÉ b·∫≠t/t·∫Øt l∆∞·ªõi **ri√™ng cho tr·ª•c X ho·∫∑c tr·ª•c Y**:


````python
plt.plot(x, y)
plt.grid(axis='x', color='red', linestyle=':', linewidth=1)  # ch·ªâ l∆∞·ªõi tr·ª•c X
plt.grid(axis='y', color='blue', linestyle='-.', linewidth=1) # ch·ªâ l∆∞·ªõi tr·ª•c Y
plt.show()
````

* `axis='both'` ‚Üí m·∫∑c ƒë·ªãnh, l∆∞·ªõi hi·ªÉn th·ªã c·∫£ hai tr·ª•c


**4. K·∫øt h·ª£p Grid v·ªõi Labels v√† Title**

````python
plt.plot(x, y, marker='o', linestyle='--', color='green')
plt.title("V√≠ d·ª• Grid trong Matplotlib")
plt.xlabel("Tr·ª•c X")
plt.ylabel("Tr·ª•c Y")
plt.grid(True, linestyle=':', color='gray', linewidth=0.7, alpha=0.7)
plt.show()
````

* K·∫øt qu·∫£: ƒë∆∞·ªùng ƒë·ª©t m√†u xanh, c√°c ƒëi·ªÉm marker tr√≤n, l∆∞·ªõi x√°m nh·∫π gi√∫p ƒë·ªçc gi√° tr·ªã d·ªÖ d√†ng.


**5. T√≥m t·∫Øt**

| Tham s·ªë                | M·ª•c ƒë√≠ch                               |
| ---------------------- | -------------------------------------- |
| `plt.grid(True/False)` | B·∫≠t ho·∫∑c t·∫Øt l∆∞·ªõi                      |
| `color`                | M√†u l∆∞·ªõi                               |
| `linestyle`            | Ki·ªÉu ƒë∆∞·ªùng l∆∞·ªõi (`-`, `--`, `-.`, `:`) |
| `linewidth`            | ƒê·ªô d√†y ƒë∆∞·ªùng l∆∞·ªõi                      |
| `alpha`                | ƒê·ªô trong su·ªët l∆∞·ªõi                     |
| `axis`                 | Ch·ªâ ƒë·ªãnh tr·ª•c (`x`, `y`, `both`)       |



**L∆∞u √Ω:**

* Grid l√† **th√†nh ph·∫ßn ph·ª• tr·ª£**, kh√¥ng l√†m thay ƒë·ªïi d·ªØ li·ªáu nh∆∞ng **tƒÉng kh·∫£ nƒÉng ƒë·ªçc bi·ªÉu ƒë·ªì**.
* N√™n d√πng grid **khi bi·ªÉu ƒë·ªì c√≥ nhi·ªÅu d·ªØ li·ªáu ho·∫∑c c·∫ßn so s√°nh gi√° tr·ªã**.


### 3.7 Matplotlib Subplot

**Subplot** trong Matplotlib cho ph√©p b·∫°n **chia m·ªôt Figure th√†nh nhi·ªÅu v√πng**, m·ªói v√πng ch·ª©a m·ªôt bi·ªÉu ƒë·ªì ri√™ng.
ƒêi·ªÅu n√†y c·ª±c k·ª≥ h·ªØu √≠ch khi so s√°nh d·ªØ li·ªáu ho·∫∑c tr√¨nh b√†y nhi·ªÅu g√≥c nh√¨n trong c√πng m·ªôt b√°o c√°o.

**1. C√°ch d√πng subplot() c∆° b·∫£n**

C√∫ ph√°p:

```python
plt.subplot(nrows, ncols, index)
```

C√∫ ph√°p:

```python
fig, ax = plt.subplots(nrows, ncols)
```

* `fig` ‚Üí ƒë·ªëi t∆∞·ª£ng Figure
* `ax` ‚Üí m·∫£ng ch·ª©a nhi·ªÅu ƒë·ªëi t∆∞·ª£ng tr·ª•c (Axes)
* D·ªÖ d√†ng t√πy ch·ªânh t·ª´ng bi·ªÉu ƒë·ªì

**V√≠ d·ª•: 2x2 bi·ªÉu ƒë·ªì**

````python
fig, ax = plt.subplots(2, 2)

ax[0, 0].plot([1,2,3], [1,2,3])
ax[0, 0].set_title("Bi·ªÉu ƒë·ªì 1")

ax[0, 1].plot([1,2,3], [3,2,1])
ax[0, 1].set_title("Bi·ªÉu ƒë·ªì 2")

ax[1, 0].bar([1,2,3], [5,3,1])
ax[1, 0].set_title("Bi·ªÉu ƒë·ªì 3")

ax[1, 1].scatter([1,2,3], [3,6,2])
ax[1, 1].set_title("Bi·ªÉu ƒë·ªì 4")

plt.tight_layout()  # t·ª± cƒÉn ch·ªânh tr√°nh tr√πng text
plt.show()
````
