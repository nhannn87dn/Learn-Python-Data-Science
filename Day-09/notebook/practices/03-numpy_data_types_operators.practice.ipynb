# Bài tập về - NumPy Data Types & Toán tử trên mảng

## **Bài 1 — Kiểm tra và chuyển đổi kiểu dữ liệu**

Cho mảng:

```python
import numpy as np
arr = np.array([10.5, 20.2, 30.7, 40.1])
```

**Yêu cầu:**

1. In ra `dtype` của mảng trên.
2. Chuyển mảng thành kiểu `int32` bằng `astype`.
3. Sau khi chuyển, hãy kiểm tra lại `dtype` mới.

## **Bài 2 — Thực hiện các phép toán số học trên mảng**

Cho hai mảng:

```python
a = np.array([5, 10, 15])
b = np.array([1, 2, 3])
```

**Yêu cầu:**
Sử dụng toán tử NumPy để tính:

* `a + b`
* `a * b`
* `a ** 2`
* `a / b`

Giải thích vì sao mỗi phép toán hoạt động được trên từng phần tử.

## **Bài 3 — Broadcasting**

Cho mảng:

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
```

**Yêu cầu:**

1. Cộng mảng `arr` với số `10` bằng broadcasting.
2. Cộng `arr` với mảng 1D `[1, 0, -1]` bằng broadcasting.
3. Giải thích vì sao NumPy cho phép hai phép cộng trên.

## **Bài 4 — Toán tử so sánh & logic**

Cho mảng:

```python
scores = np.array([65, 80, 45, 90, 50, 75])
```

**Yêu cầu:**

1. Tạo mảng boolean cho biết phần tử nào ≥ 70.
2. Tìm những điểm **vừa ≥ 50 và < 80** dùng toán tử logic (`&`, `|`).
3. Đếm xem có bao nhiêu điểm dưới 60.

## **Bài 5 — Phép toán thống kê cơ bản**

Cho mảng dữ liệu nhiệt độ theo ngày:

```python
temps = np.array([31.2, 30.5, 33.1, 34.0, 29.8, 28.7, 32.5])
```

**Yêu cầu:**

1. Tính giá trị:

   * Trung bình (`mean`)
   * Tổng (`sum`)
   * Lớn nhất (`max`)
   * Nhỏ nhất (`min`)

Sử dụng NumPy để tính.
