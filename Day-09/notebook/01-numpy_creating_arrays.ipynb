## üß© 1. T·∫°o m·∫£ng t·ª´ danh s√°ch (list)

D√πng h√†m `np.array()` ƒë·ªÉ t·∫°o m·∫£ng t·ª´ list ho·∫∑c tuple trong Python.

````python
# T·∫°o m·∫£ng 1 chi·ªÅu
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)
print("Ki·ªÉu d·ªØ li·ªáu:", type(arr1))
````

````python
# T·∫°o m·∫£ng 2 chi·ªÅu (ma tr·∫≠n)
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2)
````

### üìè Ki·ªÉm tra th√¥ng tin m·∫£ng
- `ndim`: s·ªë chi·ªÅu c·ªßa m·∫£ng
- `shape`: k√≠ch th∆∞·ªõc t·ª´ng chi·ªÅu
- `size`: t·ªïng s·ªë ph·∫ßn t·ª≠
- `dtype`: ki·ªÉu d·ªØ li·ªáu c·ªßa ph·∫ßn t·ª≠

````python
print("S·ªë chi·ªÅu:", arr2.ndim)
print("K√≠ch th∆∞·ªõc:", arr2.shape)
print("S·ªë ph·∫ßn t·ª≠:", arr2.size)
print("Ki·ªÉu d·ªØ li·ªáu:", arr2.dtype)
````

## üßÆ 2. T·∫°o m·∫£ng ƒë·∫∑c bi·ªát v·ªõi NumPy

NumPy cung c·∫•p nhi·ªÅu h√†m s·∫µn ƒë·ªÉ t·∫°o m·∫£ng nhanh:

### 2.1 `np.zeros(shape)` ‚Äì M·∫£ng to√†n s·ªë 0

````python
zeros_arr = np.zeros((2, 3))
print(zeros_arr)
````

### 2.2 `np.ones(shape)` ‚Äì M·∫£ng to√†n s·ªë 1

````python
ones_arr = np.ones((3, 2))
print(ones_arr)
````

### 2.3 `np.full(shape, value)` ‚Äì M·∫£ng v·ªõi gi√° tr·ªã c·ªë ƒë·ªãnh

````python
full_arr = np.full((2, 2), 7)
print(full_arr)
````

### 2.4 `np.eye(n)` ‚Äì Ma tr·∫≠n ƒë∆°n v·ªã

````python
eye_arr = np.eye(4)
print(eye_arr)
````

## üìà 3. T·∫°o m·∫£ng v·ªõi d√£y s·ªë

NumPy c√≥ nhi·ªÅu c√°ch ƒë·ªÉ t·∫°o c√°c d√£y s·ªë t·ª± ƒë·ªông.

### 3.1 `np.arange(start, stop, step)` ‚Äì D√£y s·ªë c√°ch ƒë·ªÅu

````python
arr_range = np.arange(0, 10, 2)
print(arr_range)
````

### 3.2 `np.linspace(start, stop, num)` ‚Äì D√£y s·ªë chia ƒë·ªÅu

Chia ƒëo·∫°n [start, stop] th√†nh `num` ph·∫ßn b·∫±ng nhau.

````python
arr_lin = np.linspace(0, 1, 5)
print(arr_lin)
````
