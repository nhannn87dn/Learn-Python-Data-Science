## üß† 1. NumPy Data Types

NumPy cung c·∫•p nhi·ªÅu **ki·ªÉu d·ªØ li·ªáu** (g·ªçi l√† `dtype`) t∆∞∆°ng t·ª± Python nh∆∞ng ƒë∆∞·ª£c t·ªëi ∆∞u cho hi·ªáu su·∫•t cao.

### ‚öôÔ∏è M·ªôt s·ªë ki·ªÉu d·ªØ li·ªáu th∆∞·ªùng g·∫∑p:

| Ki·ªÉu d·ªØ li·ªáu | M√¥ t·∫£ | V√≠ d·ª• |
|---------------|--------|--------|
| `int8`, `int16`, `int32`, `int64` | S·ªë nguy√™n c√≥ k√≠ch th∆∞·ªõc bit kh√°c nhau | `np.array([1,2,3], dtype=np.int32)` |
| `float16`, `float32`, `float64` | S·ªë th·ª±c (d·∫•u ph·∫©y ƒë·ªông) | `np.array([1.5, 2.3], dtype=np.float32)` |
| `bool` | Gi√° tr·ªã True/False | `np.array([True, False])` |
| `complex64`, `complex128` | S·ªë ph·ª©c | `np.array([1+2j, 3+4j])` |
| `str_` ho·∫∑c `U` | Chu·ªói | `np.array(["An", "B√¨nh"])` |

---
### üß© V√≠ d·ª• t·∫°o m·∫£ng v·ªõi dtype x√°c ƒë·ªãnh:

```python
# Import th∆∞ vi·ªán
import numpy as np

arr_int = np.array([1, 2, 3], dtype=np.int32)
arr_float = np.array([1.5, 2.3, 4.1], dtype=np.float64)
arr_bool = np.array([True, False, True])
arr_str = np.array(["Python", "NumPy"])

print("Ki·ªÉu d·ªØ li·ªáu arr_int:", arr_int.dtype)
print("Ki·ªÉu d·ªØ li·ªáu arr_float:", arr_float.dtype)
print("Ki·ªÉu d·ªØ li·ªáu arr_bool:", arr_bool.dtype)
print("Ki·ªÉu d·ªØ li·ªáu arr_str:", arr_str.dtype)
```

### üîÑ Thay ƒë·ªïi ki·ªÉu d·ªØ li·ªáu (`astype`)

B·∫°n c√≥ th·ªÉ **chuy·ªÉn ƒë·ªïi ki·ªÉu d·ªØ li·ªáu** c·ªßa m·∫£ng b·∫±ng ph∆∞∆°ng th·ª©c `.astype(new_type)`:

```python
arr = np.array([1.1, 2.9, 3.5])
arr_int = arr.astype(np.int32) #√âp ki·ªÉu sang int
arr_str = arr.astype(str) # √âp ki·ªÉu th√†nh str (String)

print("M·∫£ng g·ªëc:", arr)
print("Chuy·ªÉn sang int:", arr_int)
print("Chuy·ªÉn sang str:", arr_str)
```

## ‚öôÔ∏è 2. To√°n t·ª≠ tr√™n m·∫£ng (Array Operators)

NumPy cho ph√©p **th·ª±c hi·ªán to√°n t·ª≠ tr·ª±c ti·∫øp gi·ªØa c√°c m·∫£ng**, gi√∫p x·ª≠ l√Ω d·ªØ li·ªáu nhanh v√† ng·∫Øn g·ªçn.

### Bao g·ªìm:
- üßÆ To√°n t·ª≠ s·ªë h·ªçc
- üöÄ Broadcasting
- üîç To√°n t·ª≠ logic
- üß† To√°n t·ª≠ ma tr·∫≠n
- üìä Ph√©p to√°n th·ªëng k√™ c∆° b·∫£n

---
### üßÆ 2.1 To√°n t·ª≠ s·ªë h·ªçc

```python
a = np.array([10, 20, 30])
b = np.array([2, 4, 6])

print("C·ªông:", a + b)
print("Tr·ª´:", a - b)
print("Nh√¢n:", a * b)
print("Chia:", a / b)
print("L≈©y th·ª´a:", a ** b)
```

---
### üöÄ 2.2 Broadcasting

> **Broadcasting** cho ph√©p NumPy th·ª±c hi·ªán ph√©p to√°n gi·ªØa c√°c m·∫£ng c√≥ **k√≠ch th∆∞·ªõc kh√°c nhau**, b·∫±ng c√°ch **t·ª± m·ªü r·ªông** m·∫£ng nh·ªè h∆°n cho ph√π h·ª£p.

V√≠ d·ª•: m·∫£ng 1D

```python
a = np.array([1, 2, 3])
b = 2
print(a + b)
# üëâ [3 4 5]
```

·ªû ƒë√¢y, `b` ƒë∆∞·ª£c "broadcast" th√†nh `[2, 2, 2]`.
V·ªõi m·∫£ng 2 chi·ªÅu:

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])
B = np.array([10, 20, 30])

print(A + B)
# üëâ [[11 22 33]
#      [14 25 36]]
```


**Quy t·∫Øc broadcast**

* N·∫øu hai m·∫£ng c√≥ s·ªë chi·ªÅu kh√°c nhau ‚Üí th√™m chi·ªÅu 1 v√†o m·∫£ng nh·ªè h∆°n.
* N·∫øu k√≠ch th∆∞·ªõc kh√¥ng kh·ªõp, NumPy s·∫Ω c·ªë m·ªü r·ªông chi·ªÅu c√≥ k√≠ch th∆∞·ªõc 1.
* N·∫øu v·∫´n kh√¥ng kh·ªõp ‚Üí b√°o l·ªói.

---
### üîç 2.3 To√°n t·ª≠ logic

NumPy h·ªó tr·ª£ c√°c ph√©p so s√°nh ph·∫ßn t·ª≠:

| To√°n t·ª≠ | √ù nghƒ©a |
|----------|----------|
| `==` | So s√°nh b·∫±ng |
| `!=` | Kh√°c |
| `>` | L·ªõn h∆°n |
| `<` | Nh·ªè h∆°n |
| `>=`, `<=` | L·ªõn h∆°n ho·∫∑c b·∫±ng / Nh·ªè h∆°n ho·∫∑c b·∫±ng |

```python
x = np.array([1, 2, 3, 4])
y = np.array([2, 2, 3, 1])

print("x == y:", x == y)
print("x > y:", x > y)
print("x < y:", x < y)
```

---
### üß† 2.4 To√°n t·ª≠ ma tr·∫≠n

NumPy cung c·∫•p c√°c ph√©p to√°n ma tr·∫≠n nh∆∞:

- `@` ho·∫∑c `np.dot(a, b)` : nh√¢n ma tr·∫≠n
- `np.transpose(a)` : chuy·ªÉn v·ªã
- `np.linalg.inv(a)` : ngh·ªãch ƒë·∫£o (n·∫øu vu√¥ng)
- `np.linalg.det(a)` : ƒë·ªãnh th·ª©c

```python
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[5, 6], [7, 8]])

print("Nh√¢n ma tr·∫≠n:\n", m1 @ m2)
print("Chuy·ªÉn v·ªã:\n", np.transpose(m1))
print("ƒê·ªãnh th·ª©c:", np.linalg.det(m1))
```

---
### üìä 2.5 Th·ªëng k√™ c∆° b·∫£n

M·ªôt s·ªë ph√©p to√°n th·ªëng k√™ ph·ªï bi·∫øn trong NumPy:

| H√†m | M√¥ t·∫£ |
|------|-------|
| `np.sum()` | T·ªïng c√°c ph·∫ßn t·ª≠ |
| `np.mean()` | Trung b√¨nh |
| `np.median()` | Trung v·ªã |
| `np.std()` | ƒê·ªô l·ªách chu·∫©n |
| `np.max()` / `np.min()` | Gi√° tr·ªã l·ªõn nh·∫•t / nh·ªè nh·∫•t |

```python
data = np.array([5, 10, 15, 20, 25])

print("T·ªïng:", np.sum(data))
print("Trung b√¨nh:", np.mean(data))
print("Trung v·ªã:", np.median(data))
print("ƒê·ªô l·ªách chu·∫©n:", np.std(data))
print("Gi√° tr·ªã l·ªõn nh·∫•t:", np.max(data))
```
