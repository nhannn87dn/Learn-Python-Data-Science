## üß† 1. Array Shape (H√¨nh d·∫°ng m·∫£ng)

**`shape`** trong NumPy bi·ªÉu th·ªã **s·ªë chi·ªÅu** v√† **s·ªë ph·∫ßn t·ª≠ ·ªü m·ªói chi·ªÅu** c·ªßa m·∫£ng.

### üìñ C√∫ ph√°p:
```python
array.shape
```

### üìò V√≠ d·ª•:

````python
# Import th∆∞ vi·ªán
import numpy as np
````

````python
# M·∫£ng 1 chi·ªÅu
arr1 = np.array([1, 2, 3, 4, 5])
print("M·∫£ng 1D:", arr1)
print("Shape:", arr1.shape)
````

````python
# M·∫£ng 2 chi·ªÅu
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("\nM·∫£ng 2D:\n", arr2)
print("Shape:", arr2.shape)
````

````python
# M·∫£ng 3 chi·ªÅu
arr3 = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("\nM·∫£ng 3D:\n", arr3)
print("Shape:", arr3.shape)
````

## üß© 2. Reshaping Arrays (Thay ƒë·ªïi h√¨nh d·∫°ng m·∫£ng)

Ph∆∞∆°ng th·ª©c **`reshape()`** cho ph√©p thay ƒë·ªïi s·ªë chi·ªÅu ho·∫∑c k√≠ch th∆∞·ªõc c·ªßa m·∫£ng m√† **kh√¥ng l√†m thay ƒë·ªïi d·ªØ li·ªáu**.

### üìñ C√∫ ph√°p:
`numpy.reshape(array, new_shape)`
ho·∫∑c
`array.reshape(new_shape)`


üìå **L∆∞u √Ω:** T·ªïng s·ªë ph·∫ßn t·ª≠ ph·∫£i gi·ªØ nguy√™n.

V√≠ d·ª•: m·∫£ng c√≥ 6 ph·∫ßn t·ª≠ th√¨ ch·ªâ c√≥ th·ªÉ reshape th√†nh `(2,3)`, `(3,2)`, `(6,1)`... nh∆∞ng **kh√¥ng th·ªÉ** `(2,2)`.

````python
arr = np.array([1, 2, 3, 4, 5, 6])
print("M·∫£ng ban ƒë·∫ßu:", arr)
````

````python
# Reshape th√†nh 2 h√†ng, 3 c·ªôt
reshaped = arr.reshape(2, 3)
print("\nReshape (2,3):\n", reshaped)
````

````python
# Reshape th√†nh 3 h√†ng, 2 c·ªôt
reshaped2 = arr.reshape(3, 2)
print("\nReshape (3,2):\n", reshaped2)
````

### ‚ö†Ô∏è Quy t·∫Øc quan tr·ªçng:

T·ªïng s·ªë ph·∫ßn t·ª≠ tr∆∞·ªõc v√† sau **ph·∫£i b·∫±ng nhau**.

N·∫øu vi ph·∫°m quy t·∫Øc n√†y, NumPy s·∫Ω b√°o l·ªói:

````python
try:
    wrong_shape = arr.reshape(4, 2)  # 6 ph·∫ßn t·ª≠ kh√¥ng chia h·∫øt cho 8
except Exception as e:
    print("‚ùå L·ªói:", e)
````

### üí° M·∫πo: D√πng `-1` ƒë·ªÉ NumPy t·ª± t√≠nh k√≠ch th∆∞·ªõc c√≤n l·∫°i

B·∫°n c√≥ th·ªÉ truy·ªÅn `-1` cho m·ªôt chi·ªÅu, NumPy s·∫Ω t·ª± suy ra gi√° tr·ªã h·ª£p l·ªá d·ª±a tr√™n t·ªïng s·ªë ph·∫ßn t·ª≠.

````python
arr = np.arange(12)
print("M·∫£ng g·ªëc:", arr)
````

````python
# NumPy t·ª± t√≠nh chi·ªÅu c√≤n l·∫°i
reshaped_auto = arr.reshape(3, -1)
print("\nReshape (3, -1):\n", reshaped_auto)

reshaped_auto2 = arr.reshape(-1, 4)
print("\nReshape (-1, 4):\n", reshaped_auto2)
````

## üßæ 3. Flatten Array (L√†m ph·∫≥ng m·∫£ng)

Ph∆∞∆°ng th·ª©c **`flatten()`** d√πng ƒë·ªÉ **bi·∫øn m·∫£ng nhi·ªÅu chi·ªÅu th√†nh 1 chi·ªÅu**.

### üìñ C√∫ ph√°p:
```python
array.flatten()
```

### üìò V√≠ d·ª•:

````python
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
print("M·∫£ng 2D:\n", arr2d)

flat = arr2d.flatten()
print("\nSau khi flatten:", flat)
````

### üîÅ So s√°nh `ravel()` v√† `flatten()`:

| H√†m | Tr·∫£ v·ªÅ | Ghi ch√∫ |
|------|---------|----------|
| `flatten()` | B·∫£n **sao (copy)** c·ªßa m·∫£ng | Thay ƒë·ªïi b·∫£n g·ªëc **kh√¥ng ·∫£nh h∆∞·ªüng** |
| `ravel()` | B·∫£n **tham chi·∫øu (view)** | Thay ƒë·ªïi b·∫£n g·ªëc **·∫£nh h∆∞·ªüng** |

üëâ Th∆∞·ªùng d√πng `flatten()` khi mu·ªën m·∫£ng m·ªõi ƒë·ªôc l·∫≠p.

````python
arr = np.array([[1, 2], [3, 4]])

flat_copy = arr.flatten()
flat_view = arr.ravel()

flat_copy[0] = 99
flat_view[1] = 88

print("M·∫£ng g·ªëc sau khi s·ª≠a view:\n", arr)
print("Flatten (copy):", flat_copy)
print("Ravel (view):", flat_view)
````

- **`flatten()`**: L√†m ph·∫≥ng m·∫£ng th√†nh 1 chi·ªÅu (tr·∫£ v·ªÅ b·∫£n sao).
