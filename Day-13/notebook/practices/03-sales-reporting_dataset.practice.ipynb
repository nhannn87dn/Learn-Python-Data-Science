## 1. Dữ liệu đầu vào

Cấu trúc dữ liệu:

- `Order ID`: Mã đơn hàng
- `Product`: Tên hàng
- `Quantity Ordered` : Số lượng đặt mua
- `Price Each`: Đơn giá
- `Order Date`: Ngày đặt hàng
- `Purchase Address`: Địa chỉ mua hàng

```python
# combine_csvs.py
import glob
import pandas as pd

# 1) tìm tất cả file csv trong thư mục (chỉnh đường dẫn nếu cần)
#    ví dụ: "data/*.csv" hoặc "./" nếu file nằm cùng thư mục
files = glob.glob("../data/Sales-Reporting/data/*.csv")  # hoặc "./*.csv"

# 2) đọc từng file vào 1 DataFrame tạm và lưu vào list
dfs = []
for fp in files:
    try:
        # đọc file; nếu có cột ngày muốn parse thêm: parse_dates=['col_ngay']
        df = pd.read_csv(fp, low_memory=False)
    except UnicodeDecodeError:
        # nếu gặp lỗi encoding, thử encoding khác (ví dụ 'utf-8' -> 'latin1')
        df = pd.read_csv(fp, encoding="latin1", low_memory=False)
    # thêm dữ liệu vào list
    dfs.append(df)

# 3) gộp tất cả DataFrame (union các cột khác nhau)
#    sort=False giữ nguyên thứ tự cột (không bắt buộc)
combined = pd.concat(dfs, ignore_index=True, sort=False)

# 4) (tuỳ chọn) chuẩn hoá kiểu dữ liệu thông minh
combined = combined.convert_dtypes()  # chuyển sang dtype mới (Int64, string,...)

# 5) xuất ra 1 file duy nhất
combined.to_csv("combined_12months.csv", index=False)

print(f"Đã gộp {len(files)} file -> combined_12months.csv, shape: {combined.shape}")
```
